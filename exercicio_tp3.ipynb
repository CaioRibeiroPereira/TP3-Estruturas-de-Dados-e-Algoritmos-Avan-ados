# Exercício 01

```python
from collections import deque

def contar_grupos(n, amizades):
    grafo = {i: [] for i in range(1, n + 1)}

    for a, b in amizades:
        grafo[a].append(b)
        grafo[b].append(a)

    visitados = set()
    grupos = 0

    for aluno in range(1, n + 1):
        if aluno not in visitados:
            grupos += 1

            fila = deque([aluno])
            visitados.add(aluno)

            while fila:
                atual = fila.popleft()

                for vizinho in grafo[atual]:
                    if vizinho not in visitados:
                        visitados.add(vizinho)
                        fila.append(vizinho)

    return grupos



n = 6
amizades = [
    (1, 2),
    (2, 3),
    (4, 5)
]

print("Número de grupos:", contar_grupos(n, amizades))
```

**Complexidade**

A construção do grafo possui complexidade O(M), onde M é o número de amizades.

A BFS visita cada aluno uma única vez e percorre cada amizade no máximo duas vezes, resultando em complexidade O(N + M).

Portanto, a complexidade total da solução é:

O(N + M)

onde:

N = número de alunos (vértices)
M = número de amizades (arestas)

# Exercício 02

```python
from collections import deque

def contar_passeios_validos(S, tuneis, passeios):
    grafo = {i: [] for i in range(1, S + 1)}

    for a, b in tuneis:
        grafo[a].append(b)
        grafo[b].append(a)

    def conectado(origem, destino):
        fila = deque([origem])
        visitados = {origem}

        while fila:
            atual = fila.popleft()

            if atual == destino:
                return True

            for vizinho in grafo[atual]:
                if vizinho not in visitados:
                    visitados.add(vizinho)
                    fila.append(vizinho)

        return False

    validos = 0

    for passeio in passeios:
        eh_valido = True

        for i in range(len(passeio) - 1):
            if not conectado(passeio[i], passeio[i + 1]):
                eh_valido = False
                break

        if eh_valido:
            validos += 1

    return validos


S = 6

tuneis = [
    (1, 2),
    (2, 3),
    (4, 5),
    (5, 6)
]

passeios = [
    [1, 3],# válido
    [4, 6], # válido
    [1, 5],  # inválido
    [1, 2, 3]  # válido
]

print("Passeios válidos:", contar_passeios_validos(S, tuneis, passeios))
```

**Complexidade**

A construção do grafo possui complexidade O(T), onde T é o número de túneis.

Cada verificação de conectividade utiliza BFS, com complexidade O(S + T).

Sejam:

P = número de passeios
K = tamanho total das sequências avaliadas

Como para cada par consecutivo de salões pode ser executada uma BFS, a complexidade total é:

O(T + K(S + T))

onde:

S = número de salões (vértices)
T = número de túneis (arestas)
K = total de pares consecutivos verificados em todos os passeios

**Complexidade**

A inserção de uma ponte possui complexidade:

O(1)

pois apenas adiciona a aresta às listas de adjacência.

Cada consulta utiliza BFS para verificar conectividade entre duas ilhas, com complexidade:

O(N + M)

onde:

N = número de ilhas (vértices)
M = número de pontes construídas (arestas)

No pior caso, se todas as operações forem consultas, a complexidade total será:

O(M(N + M))

considerando até M operações de busca ao longo da execução.

# Exercício 05


```python
from collections import deque

def bfs(grafo, origem):
    visitados = {origem}
    fila = deque([origem])
    ordem = []

    while fila:
        atual = fila.popleft()
        ordem.append(atual)

        for vizinho in sorted(grafo[atual]):
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(vizinho)

    return ordem


grafo = {
    "brush": ["nail_polish"],
    "nail_polish": ["brush", "eye_shadow", "nails"],
    "eye_shadow": ["nail_polish", "eye_glasses"],
    "eye_glasses": ["eye_shadow"],
    "nails": ["nail_polish", "pins", "needles", "hammer"],
    "pins": ["nails", "needles"],
    "needles": ["nails", "pins"],
    "hammer": ["nails", "drill", "saw"],
    "drill": ["hammer"],
    "saw": ["hammer", "knife"],
    "knife": ["saw", "fork", "spoon"],
    "fork": ["knife"],
    "spoon": ["knife"]
}

ordem = bfs(grafo, "nails")

print("Ordem BFS:")
print(" - ".join(ordem))
```

*Comparação entre BFS e DFS*

A DFS aprofunda a busca em um caminho até não haver mais vértices a explorar, retornando apenas depois para visitar alternativas. Por isso, ela tende a descobrir vértices mais distantes antes dos mais próximos.

A BFS explora o grafo em camadas, visitando primeiro todos os vértices mais próximos da origem e somente depois avançando para níveis mais distantes.

Assim, a BFS é mais adequada quando se deseja priorizar recomendações próximas ao item inicial, enquanto a DFS é mais adequada para explorar cadeias longas de recomendações.

*Complexidade*

A BFS visita cada vértice uma única vez e percorre cada aresta uma única vez.

Portanto, sua complexidade é:

O(V + E)

onde:

V = número de vértices (produtos)
E = número de arestas (relações entre produtos)

# Exercício 06


```python
from collections import deque

def menor_caminho(grafo, origem, destino):
    fila = deque([origem])

    visitados = {origem}
    predecessor = {origem: None}

    while fila:
        atual = fila.popleft()

        if atual == destino:
            break

        for vizinho in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                predecessor[vizinho] = atual
                fila.append(vizinho)

    caminho = []
    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = predecessor[atual]

    caminho.reverse()

    return caminho, len(caminho) - 1


grafo = {
    "Idris": ["Kamil", "Talia"],
    "Kamil": ["Idris", "Lina"],
    "Lina": ["Kamil", "Sasha"],
    "Sasha": ["Lina", "Marco"],
    "Marco": ["Sasha", "Ken"],
    "Ken": ["Marco", "Talia"],
    "Talia": ["Idris", "Ken"]
}

caminho, distancia = menor_caminho(grafo, "Idris", "Lina")

print("Caminho mínimo:")
print(" - ".join(caminho))

print("\nNúmero de conexões necessárias foi de:")
print(distancia)
```

 *Justificativa*

A BFS explora o grafo por níveis. Assim, o primeiro momento em que um vértice é alcançado corresponde ao caminho com o menor número de arestas a partir da origem.

Por isso, o caminho:

Idris -> Kamil -> Lina

é o caminho mínimo entre Idris e Lina.

# Exercício 07

```python
from collections import deque

def dfs(grafo, vertice, visitados, ordem):
    visitados.add(vertice)
    ordem.append(vertice)

    for vizinho in grafo[vertice]:
        if vizinho not in visitados:
            dfs(grafo, vizinho, visitados, ordem)


def bfs(grafo, origem):
    fila = deque([origem])
    visitados = {origem}
    ordem = []

    while fila:
        atual = fila.popleft()
        ordem.append(atual)

        for vizinho in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                fila.append(vizinho)

    return ordem


grafo = {
    "Inicio": ["A", "B"],
    "A": ["C"],
    "B": ["C", "F"],
    "C": ["D"],
    "D": ["E"],
    "E": [],
    "F": ["E"]
}

ordem_dfs = []
visitados = set()

dfs(grafo, "Inicio", visitados, ordem_dfs)
ordem_bfs = bfs(grafo, "Inicio")

print("DFS:")
print(" - ".join(ordem_dfs))

print("\nBFS:")
print(" - ".join(ordem_bfs))
```

*Comparação*

A DFS aprofunda a busca em um caminho antes de retornar para explorar alternativas. Por isso, ao sair de Inicio, ela segue por A -C -D - E, antes de visitar B.

A BFS explora o grafo por níveis. Assim, visita primeiro todos os vértices diretamente alcançável a partir de Inicio (A e B), depois seus sucessores (C e F), e assim por diante.

Em um grafo direcionado, a direção das arestas limitam os caminhos possíveis. Por exemplo, existe uma aresta de Inicio - A, mas não de A - Inicio. Portanto, após chegar em A, não é possível retornar para Inicio utilizando essa aresta.

* Complexidade

Tanto a DFS quanto a BFS visitam cada vértice uma única vez e percorrem cada aresta uma única vez.

Portanto, a complexidade de ambos os algoritmos é:

O(V + E)

onde:

V = número de vértices
E = número de arestas direcionadas

# Exercício 08

```python
from collections import deque

def bfs_caminho(grafo, origem, destino):
    fila = deque([origem])

    visitados = {origem}
    predecessor = {origem: None}

    while fila:
        atual = fila.popleft()

        if atual == destino:
            break

        for vizinho, peso in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                predecessor[vizinho] = atual
                fila.append(vizinho)

    caminho = []
    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = predecessor[atual]

    caminho.reverse()

    custo = 0

    for i in range(len(caminho) - 1):
        atual = caminho[i]
        proximo = caminho[i + 1]

        for vizinho, peso in grafo[atual]:
            if vizinho == proximo:
                custo += peso
                break

    return visitados, caminho, custo


grafo = {
    "Berco_A": [("Patio_1", 4), ("Patio_2", 7)],
    "Berco_B": [("Patio_2", 3), ("Patio_3", 6)],
    "Patio_1": [("Berco_A", 4), ("Patio_2", 2), ("Alfandega", 8)],
    "Patio_2": [("Berco_A", 7), ("Berco_B", 3), ("Patio_1", 2), ("Patio_3", 2), ("Alfandega", 5)],
    "Patio_3": [("Berco_B", 6), ("Patio_2", 2), ("Centro_Logistico", 4)],
    "Alfandega": [("Patio_1", 8), ("Patio_2", 5), ("Centro_Logistico", 3)],
    "Centro_Logistico": [("Patio_3", 4), ("Alfandega", 3)]
}

alcancaveis, caminho, custo = bfs_caminho(
    grafo,
    "Berco_A",
    "Centro_Logistico"
)

print("Áreas alcançáveis:")
print(sorted(alcancaveis))

print("\nCaminho com menor número de etapas:")
print(" - ".join(caminho))

print("\nCusto total:")
print(custo)
```

*Explicação*

A BFS ignora os pesos das arestas e considera apenas o número de etapas entre os vértices.

Assim, ela encontra o caminho:

Berco_A - Patio_1 - Alfandega - Centro_Logistico

que possui apenas 3 arestas.

Entretanto, esse caminho não necessariamente possui o menor custo operacional. Um caminho com mais etapas pode ter custo total menor, pois a BFS não utiliza os pesos durante a busca.

# Exercício 09

```python
import heapq

def dijkstra(grafo, origem):
    distancias = {vertice: float("inf") for vertice in grafo}
    predecessores = {vertice: None for vertice in grafo}

    distancias[origem] = 0

    fila_prioridade = [(0, origem)]

    while fila_prioridade:
        distancia_atual, atual = heapq.heappop(fila_prioridade)

        if distancia_atual > distancias[atual]:
            continue

        for vizinho, peso in grafo[atual]:
            nova_distancia = distancias[atual] + peso

            if nova_distancia < distancias[vizinho]:
                distancias[vizinho] = nova_distancia
                predecessores[vizinho] = atual

                heapq.heappush(
                    fila_prioridade,
                    (nova_distancia, vizinho)
                )

    return distancias, predecessores


grafo = {
    "Berco_A": [("Patio_1", 4), ("Patio_2", 7)],
    "Berco_B": [("Patio_2", 3), ("Patio_3", 6)],
    "Patio_1": [("Berco_A", 4), ("Patio_2", 2), ("Alfandega", 8)],
    "Patio_2": [("Berco_A", 7), ("Berco_B", 3), ("Patio_1", 2), ("Patio_3", 2), ("Alfandega", 5)],
    "Patio_3": [("Berco_B", 6), ("Patio_2", 2), ("Centro_Logistico", 4)],
    "Alfandega": [("Patio_1", 8), ("Patio_2", 5), ("Centro_Logistico", 3)],
    "Centro_Logistico": [("Patio_3", 4), ("Alfandega", 3)]
}

distancias, predecessores = dijkstra(grafo, "Berco_A")

print("Menores distâncias a partir de Berco_A:\n")

for vertice, distancia in distancias.items():
    print(f"{vertice}: {distancia}")
```

```python
from collections import deque
import heapq


def bfs(grafo, origem, destino):
    fila = deque([origem])

    visitados = {origem}
    predecessor = {origem: None}

    while fila:
        atual = fila.popleft()

        if atual == destino:
            break

        for vizinho, peso in grafo[atual]:
            if vizinho not in visitados:
                visitados.add(vizinho)
                predecessor[vizinho] = atual
                fila.append(vizinho)

    caminho = []
    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = predecessor[atual]

    caminho.reverse()

    custo = 0

    for i in range(len(caminho) - 1):
        for vizinho, peso in grafo[caminho[i]]:
            if vizinho == caminho[i + 1]:
                custo += peso
                break

    return caminho, custo


def dijkstra(grafo, origem):
    distancias = {v: float("inf") for v in grafo}
    predecessores = {v: None for v in grafo}

    distancias[origem] = 0

    fila_prioridade = [(0, origem)]

    while fila_prioridade:
        distancia_atual, atual = heapq.heappop(fila_prioridade)

        if distancia_atual > distancias[atual]:
            continue

        for vizinho, peso in grafo[atual]:
            nova_distancia = distancias[atual] + peso

            if nova_distancia < distancias[vizinho]:
                distancias[vizinho] = nova_distancia
                predecessores[vizinho] = atual

                heapq.heappush(
                    fila_prioridade,
                    (nova_distancia, vizinho)
                )

    return distancias, predecessores


def reconstruir_caminho(predecessores, origem, destino):
    caminho = []
    atual = destino

    while atual is not None:
        caminho.append(atual)
        atual = predecessores[atual]

    caminho.reverse()

    return caminho


grafo = {
    "Berco_A": [("Patio_1", 4), ("Patio_2", 7)],
    "Berco_B": [("Patio_2", 3), ("Patio_3", 6)],
    "Patio_1": [("Berco_A", 4), ("Patio_2", 2), ("Alfandega", 8)],
    "Patio_2": [("Berco_A", 7), ("Berco_B", 3), ("Patio_1", 2), ("Patio_3", 2), ("Alfandega", 5)],
    "Patio_3": [("Berco_B", 6), ("Patio_2", 2), ("Centro_Logistico", 4)],
    "Alfandega": [("Patio_1", 8), ("Patio_2", 5), ("Centro_Logistico", 3)],
    "Centro_Logistico": [("Patio_3", 4), ("Alfandega", 3)]
}

caminho_bfs, custo_bfs = bfs(
    grafo,
    "Berco_A",
    "Centro_Logistico"
)

distancias, predecessores = dijkstra(
    grafo,
    "Berco_A"
)

caminho_dijkstra = reconstruir_caminho(
    predecessores,
    "Berco_A",
    "Centro_Logistico"
)

print("Caminho BFS:")
print(" - ".join(caminho_bfs))
print("Custo:", custo_bfs)

print("\nCaminho Dijkstra:")
print(" - ".join(caminho_dijkstra))
print("Custo:", distancias["Centro_Logistico"])
```
